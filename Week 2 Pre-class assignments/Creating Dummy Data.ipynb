**[Problem 1] Creating random numbers**

```python
import numpy as np
__init__ = [1 , 0.8 , 0.8 , 1.0] #initial cov matrix
mean = (3,0)
cov = np.array(__init__).reshape(2,2)
print(cov)
list_rd = np.random.multivariate_normal(mean, cov, 500)
list_rd.reshape(500,2)
print(list_rd)
np.random.seed(0)
```

```python
print(len(list_rd))
```

**[Problem 2] Visualization by scatter plot**

```python
import matplotlib.pyplot as plt
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(list_rd[:,0],list_rd[:,1])
plt.show()
```

**[Problem 3] Visualization by histogram**

```python
import matplotlib.pyplot as plt
plt.title("histogram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(list_rd[:,1],rwidth=1)
plt.xlim(-3,6)
plt.show()
plt.title("histogram of x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(list_rd[:,0])
plt.xlim(-3,6)
plt.show()
```
