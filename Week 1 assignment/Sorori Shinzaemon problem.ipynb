> "When Hideyoshi offered him a reward, Shinzaemon was asked what he would like to receive, and he asked for one grain of rice today, two grains the next day, four grains the day after that, and so on, doubling the amount of rice each day for 100 days."

The parttern of this problem is that the number of rice today is double in compare to that of yesterday. If we regard "today" mentioned above as day one, we can easily come to the formular representing the number of rice after n day(s) - Sn:

$$S_{n} = 2^0+2^1+2^2+...2^{n-1} $$

【Problem 1] Number of grains of rice on the 100th day

### What I will do:


* Create a list named N (contain the number of rice each day) and a list named T (contain the total of rice accumulated since day one) to make the graph.
* Insert value to list N,T by using for function & the formular above.
* Calculate the totoal number of rice recieved 
* Plot the graph.

```python
N=[] #Number of rice each day
T=[] #Total number of accumulated since day one 
S=0  #Innitialize S
for i in range(0,100): #Since the list start from N[0] counter i will run from 0 to 99 represent the power in the formular
    S += 2**i #S presents the accumulated rice
    N.append(2**i); 
    T.append(S)
"""
Since the for-loop stops at i=99, 
the current value of S will be the total number of rice grains Shinzaemon-san would receive from day 1 to day 100.
"""
print("The number of rice after 100 days is:{} grains".format(S))     
```

> Wow, that's a lot. No wonder Hideyoshi-san couldn't afford these.

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Initalize a Figure and Axes
fig, ax = plt.subplots()
plt.title("The change in the number of rice each day")
plt.xlabel("Number of day")
plt.ylabel("Number of rice")
plt.plot(N,label="The amount of rice received that day") # Enter the variable name of the list in "List name"
plt.plot(T,label="The total ammount accumulated")
plt.legend(bbox_to_anchor=(1.05,1),loc = 'best', borderaxespad=0.)
plt.show()
```
